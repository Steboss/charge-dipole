```python
%pylab inline
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np
import math
import re
import os
import seaborn as sbn
sbn.set_style("whitegrid")

import matplotlib.image as mpimg
import glob
```

```python
def histogram(ifile):
    
    #collect values
    values = []
    for val in ifile.readlines():
        values.append(float(val))
  
    hist,bins = np.histogram(values,bins=20)#50)
    width = 0.7*(bins[1] - bins[0])
    center = (bins[:-1] + bins[1:])/2

    return hist,center,width
```

```python
three = open("3_test.dat","r")
ten = open("10_test.dat","r")
twe = open("20_test.dat","r")

hist3,center3,width3 = histogram(three)
hist10,center10,width10 = histogram(ten)
hist20,center20,width20 = histogram(twe)
```

```python
color = sbn.color_palette(n_colors=8)
#create subplots for  RF-Ion 30 / 50 / 80  and  RF-Water 30/ 50 /80
fig,ax = plt.subplots(1,3,figsize=(20,10))

ax[0].bar(center3,hist3,align="center",width=width3)
ax[1].bar(center10,hist10,align="center",width=width10)
ax[2].bar(center20,hist20,align="center",width=width20)
```
