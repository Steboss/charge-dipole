```python
%pylab inline
import matplotlib.pyplot as plt
import matplotlib as mpl

import numpy as np
import math
import re
import os
import seaborn as sbn
sbn.set_style("whitegrid")

import matplotlib.image as mpimg
import glob
```

```python
def extract_results(ifiles):
    degs = []
    results = {}
    for ifile in ifiles:
        inpt = open(ifile,"r").readlines()  
        deg = int(ifile.split("/")[2].split(".")[0])
        degs.append(deg)
        results[deg] = [ [], [] ] 
        for line in inpt:
            dist = float(line.split(",")[0])
            nrg = float(line.split(",")[1])
            results[deg][0].append(dist)
            results[deg][1].append(nrg)
            
    return results
```

```python
#Take theory, scale1-5-1-20
theory_files = glob.glob("results/theory/*.csv")
theory_dict = extract_results(theory_files)
#scale 1
scale_1_files = glob.glob("results/scale_1/*.csv")
scale_1 = extract_results(scale_1_files)
#scale 5
scale_5_files = glob.glob("results/scale_5/*.csv")
scale_5 = extract_results(scale_5_files)
#scale 10
scale_10_files = glob.glob("results/scale_10/*.csv")
scale_10 = extract_results(scale_10_files)
#scale 20
scale_20_files = glob.glob("results/scale_20/*.csv")
scale_20 = extract_results(scale_20_files)
```

```python
#now plot 
#first subdivid iinto  different plots to see the general trend for the first four scalings
#Comparison between all the curves
color = sbn.color_palette(n_colors=20)
#create subplots for  RF-Ion 30 / 50 / 80  and  RF-Water 30/ 50 /80
fig,ax = plt.subplots(2,2,figsize=(20,10))

#Theory
cc = 0 
for key in theory_dict:
    ax[0][0].plot(theory_dict[key][0], theory_dict[key][1],color=color[cc],marker="o",label="%d" % key) 
    ax[0][0].set_xlabel("r/$\AA$",fontsize=20)
    ax[0][0].tick_params(labelsize=20)
    ax[0][0].set_ylabel(r"$\left \langle V(r) \right \rangle$ / kcal$\cdot$mol$^{-1}$" ,fontsize=20) 
    ax[0][0].set_xlim(5,60)
    ax[0][0].set_title("Theory")
    #ax[0][0].set_ylim(-0.1,0.1)
    cc+=1 #update the color
ax[0][0].legend(loc="upper center",bbox_to_anchor=(0.5, 1.0), ncol=7, fancybox=True, shadow=True,fontsize=10)

#computed scale 1
cc = 0 
for key in theory_dict:
    ax[0][1].plot(scale_1[key][0], scale_1[key][1],color=color[cc],marker="o",label="%d" % key) 
    ax[0][1].set_xlabel("r/$\AA$",fontsize=20)
    ax[0][1].tick_params(labelsize=20)
    ax[0][1].set_ylabel(r"$\left \langle V(r) \right \rangle$ / kcal$\cdot$mol$^{-1}$" ,fontsize=20) 
    ax[0][1].set_xlim(5,60)
    ax[0][1].set_title("Scale 1")
    #ax[0][0].set_ylim(-0.1,0.1)
    cc+=1 #update the color
ax[0][1].legend(loc="upper center",bbox_to_anchor=(0.5, 1.0), ncol=7, fancybox=True, shadow=True,fontsize=10)

#computed scale 10
cc = 0 
for key in theory_dict:
    ax[1][0].plot(scale_10[key][0], scale_10[key][1],color=color[cc],marker="o",label="%d" % key) 
    ax[1][0].set_xlabel("r/$\AA$",fontsize=20)
    ax[1][0].tick_params(labelsize=20)
    ax[1][0].set_ylabel(r"$\left \langle V(r) \right \rangle$ / kcal$\cdot$mol$^{-1}$" ,fontsize=20) 
    ax[1][0].set_xlim(5,60)
    ax[1][0].set_title("Scale 10")
    #ax[0][0].set_ylim(-0.1,0.1)
    cc+=1 #update the color
ax[1][0].legend(loc="lower center",bbox_to_anchor=(0.5, 0.0), ncol=7, fancybox=True, shadow=True,fontsize=10)


#computed scale 20
cc = 0 
for key in theory_dict:
    ax[1][1].plot(scale_20[key][0], scale_20[key][1],color=color[cc],marker="o",label="%d" % key) 
    ax[1][1].set_xlabel("r/$\AA$",fontsize=20)
    ax[1][1].tick_params(labelsize=20)
    ax[1][1].set_ylabel(r"$\left \langle V(r) \right \rangle$ / kcal$\cdot$mol$^{-1}$" ,fontsize=20) 
    ax[1][1].set_xlim(5,60)
    ax[1][1].set_title("Scale 20")
    #ax[0][0].set_ylim(-0.1,0.1)
    cc+=1 #update the color
ax[1][1].legend(loc="lower center",bbox_to_anchor=(0.5, 0.0), ncol=7, fancybox=True, shadow=True,fontsize=10)

plt.tight_layout()
```
