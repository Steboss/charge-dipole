```python
%pylab inline
import matplotlib.pyplot as plt
import matplotlib as mpl
from numpy import *
import math
import re
import os
import seaborn as sbn
sbn.set_style("whitegrid")

import matplotlib.image as mpimg
import glob
```

```python
def extract_potential(folders):
    
    results = {}

    for ifolder in folders:
        #e.g.  l_0.56,0_deg/computed.csv
        #file computed.csv
        #extract the angle
        distance = []
        pots = []
        deg = ifolder.split("/")[1].split("_")[0]
        #print(deg)
        reader = open(ifolder,"r").readlines()
        for read in reader:
            r = float(read.split(",")[0])
            pot = float(read.split(",")[1])
            distance.append(r)
            pots.append(pot)
            
        results[deg] = [ distance, pots ]
        #so results[deg][0] wil be distances and results[deg][1] potential
        
    
    return results
```

```python
#Rotationally-averaged charge-dipole potential in a reaction field
#Functional form:
#(-4/3*kT)*(qu/4pieps0*r**2)**2 * ( (esp1 + esp2)**2 + (r**3/a**3)*eps1*(eps1*(r**3/a**3 - 2) - 2*eps2) ) 
l0596  = extract_potential(glob.glob("l_0.586/*_deg/computed.dat"))
theory0596 = extract_potential(glob.glob("l_0.586/*_deg/theory.dat"))


#plot
color = sbn.color_palette()
fig, ax = plt.subplots(figsize=(10,5))
ax.plot(l0596['0'][0],l0596['0'][1],color=color[0],marker="o",label="Computed - 0 deg")
ax.plot(theory0596['0'][0],theory0596['0'][1],color=color[0],marker="o",label="Theory - 0 deg")

ax.plot(l0596['30'][0],l0596['30'][1],color=color[1],marker="o",label="Computed - 30 deg")
ax.plot(theory0596['30'][0],theory0596['30'][1],color=color[1],marker="o",label="Theory - 30 deg")

ax.plot(l0596['60'][0],l0596['60'][1],color=color[2],marker="o",label="Computed - 60 deg")
ax.plot(theory0596['60'][0],theory0596['60'][1],color=color[2],marker="o",label="Theory - 60 deg")

ax.set_xlabel("r / $\AA$",fontsize=20)
ax.tick_params(labelsize=18)
ax.legend(loc="best",fontsize=20)
ax.set_ylabel(r"$\left \langle V(r) \right \rangle$ / kcal$\cdot$mol$^{-1}$" ,fontsize=20)
#ax.set_xlim(2.15,3)
ax.set_ylim(-10,10)

#plt.savefig("RFComparison_detail.png",dpi=300,transparent=True)

#compute the differences between the two curves, once the difference is less then 0.01kcal/mol print out the distsance

#for i,pots in enumerate(l0596['60'][1],0):
#    diff= pots -  theory0596['60'][1][i]
#    if abs(diff)<0.01:#
#        print(l0596['60'][0][i])
```
