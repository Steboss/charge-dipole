```python
%pylab inline
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np
import math
import re
import os
import glob
import seaborn as sbn
sbn.set_style("whitegrid")

import matplotlib.image as mpimg
from matplotlib.mlab import griddata
from scipy import interpolate
```

```python
#from r_theta.csv extract the value of bondlength L, distance r and N<DU> 
#report ech N<DU> as kT fraction , so divide by kT 
#create a heatmap with  x = r, y = L , colormap N<DU>/kT
ifile = open("r_theta.csv")
bondlength = [] #store dipole bond lengths
distance = [] #store ion-dipole distances
ndiffs = [] # store the N<DU> diff
counter = 0 

dict_average = {} #store in the dictionar, for each specific bond length the value of distance, N<DU> and std.dev
avg_list = []
tmp = 0.0
keys = []

for lines in ifile.readlines():
    bond = float(lines.split(",")[0])
    dist = float(lines.split(",")[1])
    ndiff = float(lines.split(",")[2])/(0.5921868)
    ndiffstd = float(lines.split(",")[3])/(0.5921868)
    
    
    if tmp==0:
        tmp = bond
        #keys.append(bond)
        dict_average[tmp]=[]
        dict_average[tmp].append([dist,ndiff,ndiffstd])
        
    #deal with last element of the file
    elif tmp==3.0000 and dist ==30.0000:
        #keys.append(tmp)
        dict_average[tmp].append([dist,ndiff,ndiffstd])
        
        
    elif tmp!=bond:
        tmp=bond
        dict_average[tmp]=[]
        dict_average[tmp].append([dist,ndiff,ndiffstd])
        
    else:
        dict_average[tmp].append([dist,ndiff,ndiffstd])
    

    if ndiff>1.5:
        pass
    else:
        bondlength.append(bond)
        distance.append(dist)
        ndiffs.append(ndiff)
```

```python
bondlength = np.asarray(bondlength) #L
distance = np.asarray(distance) #r
ndiffs = np.asarray(ndiffs)    #DU
#now create a meshgrid
xi,yi = np.linspace(distance.min(),distance.max(),100),np.linspace(bondlength.min(),bondlength.max(),100)
xi,yi = np.meshgrid(xi,yi)
#interpolation
zi = interpolate.griddata((distance,bondlength),ndiffs,(xi,yi),method="linear")


fig,ax = plt.subplots(figsize=(15,10))
cmap="plasma"
cax = ax.imshow(zi, vmin=ndiffs.min(), vmax=ndiffs.max(), origin='lower',
           extent=[distance.min(), distance.max(), bondlength.min(), bondlength.max()],aspect="auto",\
          cmap=cmap )

ax.set_xlabel(r"r / $\AA$",fontsize=20)
ax.set_ylabel(r"L / $\AA$",fontsize=20)
ax.tick_params(axis='x', labelsize=20)
ax.tick_params(axis='y', labelsize=20)

ticks = np.linspace(min(ndiffs),max(ndiffs),10)
cbar = fig.colorbar(cax, ticks=ticks)

#cbar.ax.set_yticklabels(['< -1', '0', '> 1'])  # vertically oriented colorbar
cbar.set_clim(0.0,1.5)
cbar.ax.set_ylabel(r"N$\langle \Delta U \rangle$ / k$_B$T",fontsize=20)
cbar.ax.tick_params(labelsize=20)
plt.savefig("heatmap_differences.png",dpi=300)
```
